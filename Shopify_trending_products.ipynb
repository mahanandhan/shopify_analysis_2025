```python
from sqlalchemy import create_engine

# Step 1: Connect to PostgreSQL
# Replace placeholders with your actual details
username = "postgres"      # default user
password = "1234" # the password you set during installation
host = "localhost"         # if running locally
port = "5432"              # default PostgreSQL port
database = "shopify_trending_products_2025"    # the database you created in pgAdmin

engine = create_engine(f"postgresql+psycopg2://{username}:{password}@{host}:{port}/{database}")

# Step 2: Load DataFrame into PostgreSQL
table_name = "shopify"   # choose any table name
df.to_sql(table_name, engine, if_exists="replace", index=False)

print(f"Data successfully loaded into table '{table_name}' in database '{database}'.")
```

```python
# Remove the dollar sign
df['price_range_usd'] = df['price_range_usd'].str.replace('$', '', regex=False)

# Split the range into lower and upper price
df[['price_min', 'price_max']] = df['price_range_usd'].str.split('-', expand=True)

# Convert to float
df['price_min'] = df['price_min'].astype(float)
df['price_max'] = df['price_max'].astype(float)

# Create average price
df['avg_price_usd'] = (df['price_min'] + df['price_max']) / 2
df.head()
```
